```python
# Даалгавар 1
import pandas as pd
import numpy as np
df = pd.read_excel("Бие даалт - 2.xlsx", index_col=0, header=0, sheet_name="Оюутнуудын нэмэлт мэдээлэл")
df = df.dropna(axis = 1, how = 'all')
df['a']= df['Оюутны код'].str[1:-6]
def kurs(x):
    if x>='19':
        return "1"
    elif x>='18':
        return "2"
    elif x>='17':
        return "3"
    else:
        return "4"
df['Курс'] = df['a'].apply(kurs)
df.pop("a")
df
```

```python
# Даалгавар 2
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
c = pd.read_excel("Бие даалт - 2.xlsx", index_col=0, header=0, sheet_name="Шалгалт")
d = pd.read_excel("Бие даалт - 2.xlsx", index_col=0, header=0, sheet_name="Ирц")
x=d.iloc[:,-1]
y=c.iloc[:,-1]
x=sm.add_constant(x)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
# Даалгавар 10
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
x1= pd.read_excel("Бие даалт - 2.xlsx", index_col=0, header=0, sheet_name="Оюутнуудын нэмэлт мэдээлэл")
y1= pd.read_excel("Бие даалт - 2.xlsx", index_col=0, header=0, sheet_name="Оюутнуудын нэмэлт мэдээлэл")
x = np.array(x1['ЭЕШ оноо']).reshape((-1, 1))
y = np.array(y1['Голч дүн']).reshape((-1, 1))
reg = LinearRegression().fit(x, y)
y_pred = reg.predict(x)
y_pred = reg.intercept_ + reg.coef_ * x
y_pred = pd.DataFrame(y_pred, columns = ['Таамагласан'])
y1['Таамаглал'] = y_pred
y1.pop("Unnamed: 1")
y1
```
