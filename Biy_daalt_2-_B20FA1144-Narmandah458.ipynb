```python
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
c1=pd.read_excel('/Users/pubee/Library/Containers/com.microsoft.Excel/Data/Downloads/Бие даалт - 2.xlsx', sheet_name='Нийт')
import matplotlib.pyplot as plt
import numpy as np



```

```python
# 1
Kurs1=pd.read_excel('/Users/pubee/Library/Containers/com.microsoft.Excel/Data/Downloads/Бие даалт - 2.xlsx', sheet_name='Оюутнуудын нэмэлт мэдээлэл')
q1=Kurs1['Оюутны код'].tolist()
def second_let(x):
    lst= []
    for wrd in x:
        lst.append(int(wrd[2]))
    return lst
second_let(q1)
ser1=pd.Series(second_let(q1))
ser2=12-ser1
Kurs1['Unnamed: 8']=pd.Series.replace(ser2)
Kurs1= Kurs1.rename(columns={'Unnamed: 8': 'Курс'})
Kurs1
```

```python
# 3-р даалгавар
Hiceel3=np.array(Kurs1['Судалсан хичээлийн тоо']).reshape((-1, 1))
Golch3=np.array(Kurs1['Голч дүн'])
reg3=LinearRegression().fit(Hiceel3, Golch3)
print('intercept:', reg3.intercept_)
print('slope:', reg3.coef_)
```

```python
# 4-р даалгавар
Eysh4=np.array(Kurs1['ЭЕШ оноо']).reshape((-1, 1))
Golch4=np.array(Kurs1['Голч дүн'])
reg2=LinearRegression().fit(Eysh4, Golch4)
print('intercept:', reg2.intercept_)
print('slope:', reg2.coef_)

```

```python
# 6-р даалгавар
Niit6=c1['Нийт'].drop(28,axis=0)
Niit6
fig = plt.figure()
# plt.title('Дүнгүүд')
ax = fig.add_axes([0,0,1,1])
langs=Kurs1['Оюутны нэр'].tolist()
percents=Niit6.tolist()
ax.bar(langs,percents)
plt.xticks(rotation=90)
plt.title('Дүнгүүд-bar chart')
plt.show()
```

```python
# 7-р даалгавар
x= Kurs1['Оюутны нэр'].tolist()
y= Niit6.tolist()
plt.title('Дүнгүүд-line chart')
plt.plot(x,y) 
plt.show() 
```

```python
# 8-р даалгавар
a1=0
a2=0
a3=0
a4=0
a5=0
for i in y:
    if i>90:
        a1+=1
    elif 90>i and i>80:
        a2+=1
    elif 80>i and i>70:
        a3+=1 
    elif 70>i and i>60:
        a4+=1
    else:
        a5+=1
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.axis('equal')
langs2 =['A','B','C','D','F']
percents2=[a1,a2,a3,a4,a5]
ax2.pie(percents2, labels = langs2,autopct='%1.2f%%')
plt.title('Дүнгүүд дугуй диаграммаар ')
plt.show()
```
